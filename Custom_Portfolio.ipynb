```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt


#cleaning data
whale_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/whale_returns.csv')
whale = pd.read_csv(whale_path,index_col="Date",infer_datetime_format=True, parse_dates=True)
whale.dropna(inplace=True)

algo_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/algo_returns.csv')
algo = pd.read_csv(algo_path,index_col="Date",infer_datetime_format=True, parse_dates=True)
algo.dropna(inplace=True)

sp500_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/sp500_history.csv') 
sp500 = pd.read_csv(sp500_path,index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500.dtypes
sp500['Close'] = sp500['Close'].str.lstrip('$')
sp500['Close'] = sp500['Close'].astype(float)
sp500 = sp500.pct_change()
sp500.dropna(inplace=True)
sp500_DR = sp500.rename(columns={'Close':'SP500 Daily Returns'})

# Concatenate all DataFrames into a single DataFrame
appended_data = pd.concat([whale,algo,sp500_DR], axis="columns",join="inner")

#new stocks, reading and cleaning
DIS_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/DIS_Returns_edit.csv')
DIS = pd.read_csv(DIS_path,index_col="Date",infer_datetime_format=True, parse_dates=True)

QIWI_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/qiwi_updated_best.csv')
QIWI = pd.read_csv(QIWI_path,index_col="Date",infer_datetime_format=True, parse_dates=True)

WING_path = Path('C:/Users/johnh/Desktop/PREWORK_JSH/Module-3/PORTFOLIO_HW/WING _returns_editnew.csv')
WING = pd.read_csv(WING_path,index_col="Date",infer_datetime_format=True, parse_dates=True)

#CONCAT
johnfolio = pd.concat([WING,QIWI,DIS], axis="columns",join="inner")
johnfolio.dropna(inplace=True)

#weighted returns
weights = [1/3, 1/3, 1/3]
johnfolio = johnfolio.dot(weights)
johnfolio = johnfolio.to_frame()
johnfolio.rename(columns={0:'Johns Returns'},inplace =True)
johnfolio = johnfolio.pct_change()


#concat appended returns and johnfolio_returns
combinedport = pd.concat([appended_data,johnfolio],axis="columns",join="inner")
combinedport.dropna(inplace=True)
combinedport

#compare risk
boxplot_updated = combinedport.boxplot(figsize=(19,6),column=['Johns Returns','SOROS FUND MANAGEMENT LLC','TIGER GLOBAL MANAGEMENT LLC','BERKSHIRE HATHAWAY INC','Algo 1','Algo 2','SP500 Daily Returns'])

#plotting daily and cumulative returns
combinedport.plot()

cm_combinedport = (1+ combinedport).cumprod()-1
cm_combinedport.plot()

#daily std
volatilitynew = combinedport.std()
volatilitynew

#annulized std
avtnew = volatilitynew * np.sqrt(252)
avtnew.sort_values(inplace=True)
avtnew

#annulized std
avt_combinedport = volatilitynew * np.sqrt(252)
avt_combinedport.sort_values(inplace=True)

#avt_combinedport

#correlation
combinedport_corr = combinedport.rolling(window=21).corr()
combinedport_corr

#combinedport_corr_heatmap = sns.heatmap(combinedport_corr,vmin=-1.5, vmax=1.15)
#combinedport_corr_heatmap

#calculating bershire hathaway rollign Beta comparing to sp500 
rolling_covar = combinedport['Johns Returns'].rolling(window=60).cov(combinedport['SP500 Daily Returns'])
rolling_var = combinedport['SP500 Daily Returns'].rolling(window=60).var()
johns_beta_60days = rolling_covar / rolling_var
johns_beta_60days.plot(title='Johns Beta 60 day rolling avg',)

#sharpe ratios
sharperatios = ((combinedport.mean() *252) / (avtnew))
sharperatios.plot().bar()
```
